```python
import pickle
from pathlib import Path
from structures import *

PKL_PATH = "results/nvidia__nemotron-3-nano.pkl"   # change this
FIXED_PATH = "results/nvidia__nemotron-3-nano_fixed.pkl"

def load_results(filename):
    with open(filename, "rb") as f:
        return pickle.load(f)

def save_results(filename, results):
    with open(filename, "wb") as f:
        pickle.dump(results, f)

results = load_results(PKL_PATH)
len(results)

results
```

```python
def print_none_results(results):
    count = 0
    for i, r in enumerate(results):
        if r.ranking is None:
            print(
                f"results[{i}] -> model={r.model}, triad_id={r.triad_id}, permutation_id={r.permutation_id}"
            )
            count += 1
    print(f"\nTotal None results: {count}")
print_none_results(results)
```

```python
def show_result(results, idx):
    r = results[idx]
    print("Index:", idx)
    print("Model:", r.model)
    print("Triad ID:", r.triad_id)
    print("Permutation ID:", r.permutation_id)
    print("Ranking:", r.ranking)
    print("\nPrompt:\n")
    print(r.prompt)

show_result(results, 44)
```

```python
def set_manual_ranking(results, idx, ranking):
    results[idx].ranking = ranking
    print(f"Updated results[{idx}]")
    print(results[idx].ranking)

set_manual_ranking(
    results,
    0,
    {
        "DE1": 1,
        "FW1": 2,
        "DU1": 3,
    }
)
set_manual_ranking(
    results,
    2,
    {
        "DE1": 1,
        "FW1": 2,
        "DU1": 3,
    }
)
set_manual_ranking(
    results,
    9,
    {
        "DE2": 1,
        "DU3": 2,
        "FW2": 3,
    }
)
set_manual_ranking(
    results,
    44,
    {
        "DE1": 1,
        "FW3": 2,
        "DU5": 3,
    }
)
set_manual_ranking(
    results,
    46,
    {
        "DE1": 1,
        "FW3": 2,
        "DU5": 3,
    }
)
```

```python
def folder_has_none_rankings(results_dir="results"):
    found_any = False

    for path in Path(results_dir).glob("*.pkl"):
        with open(path, "rb") as f:
            results = pickle.load(f)

        none_indices = [i for i, r in enumerate(results) if r.ranking is None]

        if none_indices:
            found_any = True
            print(f"{path.name}: {len(none_indices)} None rankings")
            print("Indices:", none_indices)
            print()

    if not found_any:
        print("No None rankings found in any pkl file.")

folder_has_none_rankings("results")
```
